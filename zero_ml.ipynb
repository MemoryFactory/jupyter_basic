### 1.5.2 第2个环节：数据的收集和预处理

```python
import numpy as np
import pandas as pd
from keras.datasets import mnist # 从keras中导入MNIST数据集

# 读入训练集和测试集
(X_train_image, y_train_lable), (X_test_image, y_test_lable) = mnist.load_data()
```

```python
print('数据集张量形状：', X_train_image.shape)
print('第一个数据样本：\n', X_train_image[0]) # 特征：数字图片
```

```python
print('第一个数据样本的标签：', y_train_lable[0]) # 标签：图片对应的数字

### 
for i in range(100):
    print(y_train_lable[i], end=' ')
# print(X_train_image[3])
```

```python
from keras.utils import to_categorical # 导入keras.utils工具库的类别转换工具

X_train = X_train_image.reshape(60000, 28, 28, 1) # 增加一个维度
X_test = X_test_image.reshape(10000, 28, 28, 1) # 增加一个维度
y_train = to_categorical(y_train_lable, 10) # 转换为one-hot编码
y_test = to_categorical(y_test_lable, 10) # 转换为one-hot编码

print('训练集张量形状：', X_train.shape)
print('第一个数据标签：', y_train[0])
```

### 1.5.3 第3个环节：选择机器学习模型

```python
from keras import models # 导入keras模型，以及各种神经网络的层
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D

model = models.Sequential() # 用序贯方式建立模型
model.add(Conv2D(32, (3,3), activation='relu', # 添加二维卷积层
                 input_shape=(28,28,1))) # 指定输入数据样本张量的类型
model.add(MaxPooling2D(pool_size=(2,2))) # 添加最大池化层
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25)) # 防止过拟合
model.add(Flatten())
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax')) # softmax分类激活，输出10维分类码y' # 全连接层

# 编译模型
model.compile(optimizer='rmsprop', # 指定优化器
              loss='categorical_crossentropy', # 指定损失函数
              metrics=['accuracy']) #指定验证过程中的评估指标
```

### 1.5.4 第4个环节：训练机器，确定参数

```python
model.fit(X_train, y_train, # 指定训练特征集和训练标签集
          validation_split=0.3, # 30%训练集数据拆分成验证集
          epochs=5, # 训练轮次为5轮
          batch_size=128) # 以128为批量进行训练
```

### 1.5.5 第5个环节：超参数调试和性能优化

```python
score = model.evaluate(X_test, y_test) # 在验证集上进行模型评估
print('测试集预测准确率：', score[1])
```

```python
pred = model.predict(X_test[0].reshape(1,28,28,1)) # 预测测试集第一个数据
print(pred[0], '转换一下格式得到：', pred.argmax()) # 把one-hot编码转换为数字

import matplotlib.pyplot as plt

plt.imshow(X_test[0].reshape(28,28), cmap='Grays')

###
# print(X_test[0].reshape(28,28))
```

### 3.2.7 把数据归一化

> 归一化是按比例的线性缩放。数据归一化后，数据分布不变，但是都落入一个小的特定区间，比如0\~1或者-1\~1
>
> 常见的一个归一化公式（缩放到[0,1]）：$x' = \frac{x - \min(x)}{\max(x) - \min(x)}$

```python
def scaler(train, test): # 定义归一化函数，进行数据压缩
    min = train.min(axis=0) # 按列取最小值
    max = train.max(axis=0) # 按列取最大值
    gap = max-min
    
    train -= min
    train /= gap
    test -= min
    test /= gap
    return train,test

###
# x_norm = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data)).values
###
```

## 3.3 选择机器学习模型

### 3.3.1 确定线性回归模型

> 模型：$y = wx + b$
>
>- $w$：权重(weight)
>- $b$：偏置(bias)

### 3.3.2 假设（预测）函数 —— $h(x)$

> $y' = wx + b$ 或 $h(x) = wx + b$
>
> - $y'$：所预测出的标签，叫y-hat
> - $h(x)$：机器学习所得到的函数模型，称为假设函数(hypothesis function)

### 3.3.3 损失（误差）函数 —— $L(w,b)$

> 均方误差(MSE)损失函数：$L(w,b) = \text{MSE} = \frac{1}{2N} \sum_{(x,y) \in D} (y - h(x))^2$
>
> - $(x,y)$：样本，$x$是特征，$y$是标签
> - $h(x)$：假设函数$h(x) = wx + b$，也就是预测函数$y'$
> - $D$：包含多个样本的数据集
> - $N$：样本数量。×2是为了求梯度的时候，抵消二次方后产生的系数
> - $L$：损失函数，依赖于参数$w$和$b$

```python
def loss_function(X, y, weight, bias): # 手工定义一个均方误差函数
    y_hat = weight*X + bias
    loss = y_hat-y # 求出每一个y'和训练集中真实的y之间的差异
    cost = np.sum(loss**2)/(2*len(X))
    return cost
```

```python
print('weight=5,bias=3 => loss:', loss_function(X_train,y_train,weight=5,bias=3))
print('weight=100,bias=1 => loss:',loss_function(X_train,y_train,weight=100,bias=3))
```
